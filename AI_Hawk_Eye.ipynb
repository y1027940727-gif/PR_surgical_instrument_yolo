### ÂÅµÊ∏¨ÁµêÊûúÁ∏ΩÁµê

ÊàëÂÄëÂ∞áÂ∞çÊâÄÊúâÊ∏¨Ë©¶ÂúñÁâáÁöÑÂÅµÊ∏¨ÁµêÊûúÈÄ≤Ë°åÁ∏ΩÁµêÔºåÁµ±Ë®àÊØèÂÄãÈ°ûÂà•ÁöÑÁ∏ΩÂÅµÊ∏¨Êï∏ÈáèÔºå‰∏¶Êü•ÁúãÂÅµÊ∏¨ÁãÄÊ≥Å„ÄÇ

```python
from pathlib import Path

# Á¢∫‰øù ultralytics Â∑≤ÂÆâË£ù
!pip install ultralytics
from ultralytics import YOLO

# ËºâÂÖ•Ë®ìÁ∑¥Â•ΩÁöÑÊ®°Âûã
# Á¢∫‰øù BEST_WEIGHTS ÊåáÂêëÊ≠£Á¢∫ÁöÑË∑ØÂæë (‰æÜËá™‰∏ä‰∏ÄÂÄãË®ìÁ∑¥ÁöÑËº∏Âá∫)
BEST_WEIGHTS = "/content/yolov5/runs/detect/surgical_instrument_v1/weights/best.pt"
model = YOLO(BEST_WEIGHTS)

# Ê∏¨Ë©¶ÂúñÁâáË≥áÊñôÂ§æ
SOURCE_DIR = "/content/drive/MyDrive/instruments.v2i.yolov5pytorch/test/images"

test_dir = Path(SOURCE_DIR)
image_files = sorted(test_dir.glob("*.jpg"))

total_detections = {}
images_with_detections = 0
images_without_detections = 0

print(f"ÈñãÂßãËôïÁêÜ {len(image_files)} ÂºµÊ∏¨Ë©¶ÂúñÁâá...")

for i, img_path in enumerate(image_files):
    # Â∞çÂñÆÂºµÂúñÁâáÈÄ≤Ë°åÊé®Ë´ñÔºå‰∏ç‰øùÂ≠òÂúñÁâáÂíåtxtÔºåÂõ†ÁÇ∫‰πãÂâçÂ∑≤Á∂ì‰øùÂ≠òÈÅé‰∫Ü
    # verbose=False ÂèØ‰ª•Ê∏õÂ∞ëËº∏Âá∫Ë®äÊÅØ
    results = model.predict(source=str(img_path), imgsz=640, conf=0.25, verbose=False, save=False, save_txt=False)

    boxes = results[0].boxes

    if len(boxes) > 0:
        images_with_detections += 1
        class_ids = boxes.cls.tolist()
        for cid in class_ids:
            cname = model.names[int(cid)]
            total_detections[cname] = total_detections.get(cname, 0) + 1
    else:
        images_without_detections += 1

    # È°ØÁ§∫ÈÄ≤Â∫¶
    if (i + 1) % 50 == 0 or (i + 1) == len(image_files):
        print(f"Â∑≤ËôïÁêÜ {i + 1}/{len(image_files)} ÂºµÂúñÁâá...")

print("\n=== ÂÅµÊ∏¨ÁµêÊûúÊëòË¶Å ===")
print(f"Á∏ΩÂÖ±ËôïÁêÜÂúñÁâáÊï∏: {len(image_files)}")
print(f"ÊúâÂÅµÊ∏¨Âà∞Áâ©‰ª∂ÁöÑÂúñÁâáÊï∏: {images_with_detections}")
print(f"Êú™ÂÅµÊ∏¨Âà∞Áâ©‰ª∂ÁöÑÂúñÁâáÊï∏: {images_without_detections}")

print("\nÂêÑÈ°ûÂà•Á∏ΩÂÅµÊ∏¨Êï∏Èáè:")
if total_detections:
    for cname, count in total_detections.items():
        print(f"- {cname}: {count}")
else:
    print("Ê≤íÊúâÂÅµÊ∏¨Âà∞‰ªª‰ΩïÁâ©‰ª∂„ÄÇ")

print("\n--- Ë¶ñË¶∫ÂåñÁµêÊûú --- ")
print("Â∏∂ÊúâÈÇäÁïåÊ°ÜÁöÑÂúñÁâáÂ∑≤ÂÑ≤Â≠òËá≥Ôºö`/content/yolov5/runs/demo/instrument_test`")
print("ÊÇ®ÂèØ‰ª•Âú®Ë©≤Ë≥áÊñôÂ§æ‰∏≠Êü•ÁúãÂÅµÊ∏¨ÁµêÊûú„ÄÇ")
```
